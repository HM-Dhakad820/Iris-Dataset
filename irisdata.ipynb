```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(ds[['sepal_length',	'sepal_width',	'petal_length','petal_width']],ds[['species']],test_size =0.2)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```
